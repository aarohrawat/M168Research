```python
import networkx as nx 
import matplotlib.pyplot as plt
```

```python
def graph_info(G):
    num_nodes = G.number_of_nodes()
    num_edges = G.number_of_edges()
    avg_degree = sum(dict(G.degree()).values()) / num_nodes
    num_components = nx.number_connected_components(G)
    avg_clustering = nx.average_clustering(G)
   
    #Look at the largest component and take its diameter and avg shortest path
    largest_cc = max(nx.connected_components(G), key=len)
    subG = G.subgraph(largest_cc)
    diameter = nx.diameter(subG)
    avg_path_length = nx.average_shortest_path_length(subG)

    #Print the results
    print(f"Graph with {num_nodes} nodes and {num_edges} edges")
    print(f"Average degree: {avg_degree:.2f}")
    print(f"Number of connected components: {num_components}")
    print(f"Average clustering coefficient: {avg_clustering:.2f}")
    print(f"Diameter of largest component: {diameter}")
    print(f"Average shortest path length (largest component): {avg_path_length:.2f}")
   

def plot_graph(G, title="Ego's"):
    plt.figure(figsize=(10, 8))
    nx.draw(G, 
            node_size=30, 
            node_color='skyblue', 
            edge_color='gray', 
            with_labels=False)
    plt.title(title + " Friend Network")
    plt.show()
```

Ego 0 Network

```python
ego0G = nx.read_edgelist("ego0_data/0.edges")
graph_info(ego0G)
plot_graph(ego0G, "Ego 0:")
print(nx.average_clustering(ego0G))
```
