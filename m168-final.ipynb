```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

#Enter paths to csv files
artists_path = "artist_edges.csv"
athletes_path = "athletes_edges.csv"
companies_path = "company_edges.csv"
govs_paths = "government_edges.csv"
news_path = "new_sites_edges.csv"
politicians_path = "politician_edges.csv"
pubfigs_paths = "public_figure_edges.csv"
tvshows_paths = "tvshow_edges.csv"
```

```python
def analyze_components(G, name):
    num_components = nx.number_connected_components(G)
    largest_cc = max(nx.connected_components(G), key=len)
    size_largest_cc = len(largest_cc)

    print(f"{name}")
    print(f"Number of Connected Components: {num_components}")
    print(f"Size of Largest Connected Component: {size_largest_cc}\n")
```

# Artists

```python
artists = nx.read_edgelist(artists_path, delimiter = ",", nodetype = str)
deg_dist_art = [artists.degree(n) for n in artists.nodes()]
plt.hist(deg_dist_art, bins = 10)
print(round(stats.skew(deg_dist_art), 3))
print(round(nx.average_clustering(artists), 3))
```

```python
analyze_components(artists, "Artists")
```
