```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

# Artists

```python
artists = nx.read_edgelist("/Users/ryanquach/Downloads/artist_edges.csv",
                           delimiter = ",", nodetype = str)
deg_dist_art = [artists.degree(n) for n in artists.nodes()]
```

```python
plt.hist(deg_dist_art, bins = 100);
```

```python
round(stats.skew(deg_dist_art), 3)
```

```python
round(nx.average_clustering(artists), 3)
```
